```python
years = [2022, 2010, 2000]
nashville_population = [683622, 603465, 547847]
boston_population = [650706, 617779, 590355]
bar_width = 1
bar_positions_nashville = [year - bar_width/2 for year in years]
bar_positions_boston = [year + bar_width/2 for year in years]
plt.bar(bar_positions_nashville, nashville_population, width=bar_width, label='Nashville', color='orange')
plt.bar(bar_positions_boston, boston_population, width=bar_width, label='Boston', color='purple')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Comparison: Nashville (TN) vs. Boston (MA) (2000-2022)')
plt.xticks(years)
plt.legend()
plt.show()
```

```python
cities = ['Nashville', 'Boston', 'U.S.']
population = [71.8, 69.6, 63.1]
plt.bar(cities, population, color=['orange', 'purple', 'red'])
plt.xlabel('City')
plt.ylabel('Population (%)')
plt.title('% of Civilian Labor Force (Age 16+) Nashville vs. Boston & U.S. Average')
plt.show()
```
